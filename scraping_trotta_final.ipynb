```python
from selenium import webdriver
from random import randint
from bs4 import BeautifulSoup

import requests
import time
import csv
```

```python
url_list = [
    'https://www.metacritic.com/browse/movies/score/metascore/all/filtered/netflix', 
    'https://www.metacritic.com/browse/movies/score/metascore/all/filtered/netflix?page=1',
    'https://www.metacritic.com/browse/movies/score/metascore/all/filtered/netflix?page=2',
    'https://www.metacritic.com/browse/movies/score/metascore/all/filtered/netflix?page=3',
    'https://www.metacritic.com/browse/movies/score/metascore/all/filtered/netflix?page=4',
    'https://www.metacritic.com/browse/movies/score/metascore/all/filtered/netflix?page=5',
    'https://www.metacritic.com/browse/movies/score/metascore/all/filtered/netflix?page=6',
    'https://www.metacritic.com/browse/movies/score/metascore/all/filtered/netflix?page=7',
    'https://www.metacritic.com/browse/movies/score/metascore/all/filtered/netflix?page=8'
]
```

```python


links = []

driver = webdriver.Chrome('/Users/cosimo/Documents/python/scraping/chromedriver')

driver.delete_all_cookies()

def get_urls(my_url):
    driver.get(my_url)
    time.sleep(2)
    page = driver.page_source
    
    soup = BeautifulSoup(page, "html.parser")

    movies = soup.find_all('a', class_='title')
    

    for i in range(len(movies)):
        links.append(movies[i].get('href'))
    
    
    return None



print(len(links))

for link in url_list:
    get_urls(link)
    s = randint(1, 6)
    time.sleep(s)

driver.quit()
```

```python

driver = webdriver.Chrome('/Users/cosimo/Documents/python/scraping/chromedriver')

def get_details(url):
    driver.get(url)
    ss = randint(1, 3)
    time.sleep(ss)
    page1 = driver.page_source

    soup2 = BeautifulSoup(page1, "html.parser")
    
    full_movie = []
               
    try:
        title = soup2.find('h1').text
    except:
        title = 'Page did not respond'
    
    try:
        meta_score = soup2.find(class_='metascore_w larger movie positive').text
    except:
        meta_score = 'Not Available'
        
    try:
        meta_score = soup2.find(class_='metascore_w larger movie mixed').text
    except:
        trash1 = 'Not Available' 
    
    try:
        meta_score = soup2.find(class_='metascore_w larger movie negative').text
    except:
        trash2 = 'Not Available'

    try: 
        user_score = soup2.find(class_='metascore_w user larger movie positive').text
    except:
        user_score = 'Not Available'

    try:
        user_score = soup2.find(class_='metascore_w user larger movie mixed').text
    except:
        trash3 = 'Not Available'
    
    
    try:
        user_score = soup2.find(class_='metascore_w user larger movie negative').text
    except:
        trash4 = 'Not Available'

    
    try:
        find_director = soup2.find(class_='director')
        director = find_director.find('a').text
    except:
        director = 'Not Available'


    try:
        find_starring = soup2.find(class_='summary_cast details_section').find_all('a')

        starring2 = []


        for i in range(len(find_starring)):
            starring2.append(find_starring[i].getText())

    except:
        starring2 = 'Not Available'



    try:
        release = soup2.find(class_='release_year lighter').text
    except:
        release = 'Not Available'

    
    try:
        find_genre = soup2.find(class_='genres').find_all('span')
        genre1 = find_genre[2:]
        genre = []
    except:
        genre = 'Not Available'

    try:
        for i in range(len(genre1)):
            genre.append(genre1[i].getText())
    except:
        genre = 'Not Available'

    #print(genre)         

    
    if title == '\nThis page isnâ€™t working\n\n':
        title = 'This page is not working'
    
    
    full_movie.append(title)
    full_movie.append(release)
    full_movie.append(meta_score)
    full_movie.append(user_score)
    full_movie.append(director) 
    if starring2 != 'Not Available':
        temp = ', '.join(str(e) for e in starring2)
        full_movie.append(temp)
    else:
        starring2 = 'Not Available'
        full_movie.append(starring2)
        
    
    if genre != 'Not Available':
        temp2 = ', '.join(str(e) for e in genre)
        full_movie.append(temp2)
    else:
        genre = 'Not Available'
        full_movie.append(genre)
    
    
    
    print(full_movie)
    
    return full_movie           
               
               



def write_csv():
    # open new file for writing 
    
    file = open('metacritic_scrape.csv', 'w', newline='')
    
    # make a Python CSV writer object
    
    write_file = csv.writer(file)
    
    write_file.writerow(['Title', 'Release', 'Meta_score', 'User_score', 'Director', 'Starring', 'Genre'])
    
    for movie in links:
        movie_url = 'https://www.metacritic.com' + movie
        full_movie = get_details(movie_url)
        write_file.writerow([full_movie[0], full_movie[1], full_movie[2], full_movie[3], full_movie[4], full_movie[5], full_movie[6]])
        s = randint(1, 6)
        time.sleep(s)
        
    
    # close the file - end of function 
    file.close()
    
    # prevents error
    return None
 
# run the function - this should create a .csv file 
write_csv()
driver.quit()
```
